### 1. Handling Missing Values

````python
# Missing Values/Null Values Count
df.isna().sum().sort_values(ascending = False)
````

````python
# Handling Missing Values & Missing Value
#  Imputation of glucose with median
df['glucose'] = df['glucose'].fillna(df['glucose'].median())
print('glucose Missing After:', df['glucose'].isnull().sum())

# Impute education with the most frequent value mode
df['education'] = df['education'].fillna(df['education'].mode()[0])
print('education Missing After:', df['education'].isnull().sum())

#  Imputation of totChol with median
df['totChol'] = df['totChol'].fillna(df['totChol'].median())
print('totChol Missing After:', df['totChol'].isnull().sum())
````

````python
# Remaining all the null value rows are being droped
df = df.dropna(axis=0)
````

### 2. Handling Outliers

````python
# checking for outliers
for var in numerical_columns:
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    fig = sns.boxplot(y=df[var])
    fig.set_title('')
    fig.set_ylabel(var)
    
    plt.subplot(1, 2, 2)
    fig = sns.distplot(df[var].dropna())
    fig.set_ylabel('Number of houses')
    fig.set_xlabel(var)

    plt.show()
````

````python
# Handling Outliers & Outlier treatments
# Capping the outlier rows

# cholostrol level 
df = df[df['totChol'] <= 400]

# ciggeretes per day
df = df[df['cigsPerDay'] < 50]

# systolic blood pressure
df = df[df['sysBP'] < 200]

# Diastolic blood pressure
df = df[(df['diaBP'] > 50) & (df['diaBP'] < 120)]

# Body mass index
df = df[df['BMI'] < 35]

# heart rate
df = df[(df['heartRate'] > 50) & (df['heartRate'] < 110)]

# Glucose
df = df[(df['glucose'] > 50) & (df['glucose'] < 140)]
````

### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
# create copy of CHD_data
new_df=df.copy()
````

Systolic and Diastolic blood pressure has a high correlation that can be commonly classified into hypertension.

````python
# creating blood pressure function
def blood_presure_classification(SysBP, DiaBP):
  if (SysBP < 90) or (DiaBP < 60):
    return 0 # 'Hypotension'
  if (SysBP < 120) and (DiaBP < 80):
    return 1 # 'Optimal'
  if (SysBP < 130) or (DiaBP < 86):
    return 2 # 'Normal'
  if (SysBP < 140) or (DiaBP < 90):
    return 3 # 'PreHypertensive'
  if (SysBP > 140) and (DiaBP < 90):
    return 4 # 'IsolatedSystolicHypertension'
  if (SysBP < 160) or (DiaBP < 100):
    return 5 # 'GradeI'
  if (SysBP < 180) or (DiaBP < 110):
    return 6 # 'GradeII'
  return 7 # 'GradeIII'
````

Blood glucose levels can be used to classify the severity of Diabetes.

````python
# creating glucose function
def diabetes_grade(glucose):
  if glucose < 100:
    return 1 # 'Normal'
  if glucose < 125:
    return 2 # 'Prediabetic'
  if glucose < 200:
    return 3 # 'Mild'
  if glucose < 400:
    return 4 # 'Moderate'
  return 5 # 'Severe'   
````

````python
# Manipulate Features to minimize feature correlation and create new features
new_df['Hypertension'] = df.apply(lambda x: blood_presure_classification(x['sysBP'],x['diaBP']),axis=1)
new_df['Diabetes'] = df['glucose'].apply(lambda x: diabetes_grade(x))
new_df.head()
````

````python
# Multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
````
